## 함수로 정리하기

```python
import pandas as pd
import requests
import bs4
import time
from tqdm import tqdm
```

```python
item_code = '373340'
item_name = '유진스팩6호'
```

```python
def finance_table(item_code, item_name):
    item_list= []
    content=[]
    url_list=[]
    # last_page 찾기
    url = f'https://finance.naver.com/item/board.naver?code={item_code}'
    headers = {'user-agent':'Mozilla/5.0'}
    response = requests.get(url, headers = headers)
    html = bs4.BeautifulSoup(response.text)
    box = html.find_all('td', {'class':'pgRR'})[0]
    last_page = box.find('a')['href'].split('=')[-1]
    print('last page: ', last_page)
    
    # last_page만큼 반복 
    for i in tqdm(range(1,int(last_page)+1)):
        url = f'https://finance.naver.com/item/board.naver?code={item_code}&page={i}'
        headers = {'user-agent':'Mozilla/5.0'}
        response = requests.get(url, headers = headers)
        html = bs4.BeautifulSoup(response.text)
        table = html.find("table",{'class':"type2"})
        tbody = html.find("tbody")
        td = tbody.find_all("td",{'class':"title"})
        
        df = pd.read_html(response.text)[1]
        df = df.drop(columns='Unnamed: 6',axis=1)
        df = df.dropna(axis=0)
        item_list.append(df)
        time.sleep(0.01)
        
        for j in range(len(td)):      
            address = html.find_all('td', {'class':'title'})[j]
            address = address.find('a')['href']

            URL = 'https://finance.naver.com' + address
            url_list.append(URL)
            response1 = requests.get(URL, headers = headers)

            html2 = bs4.BeautifulSoup(response1.text)
            con = html2.select('#body')[0].text.replace('\r','')
            content.append(con) 
            time.sleep(0.01)

    df_total = pd.concat(item_list)
    df_total = df_total.drop_duplicates()
    
    df_total['내용'] = content
    df_total['링크'] = url_list
    
    # 파일 저장
    file_name = f"{item_name}_{item_code}.csv"
    
    df_total.to_csv(file_name, index=False)
    df = pd.read_csv(file_name, encoding='utf-8')
    
    return df
```

```python
df = finance_table(item_code, item_name)
df
```
